```python
from numpy import percentile

# List of columns
columns = ['movie', 'Director1', 'Duration', 'year of release', 'Rating',
           'Metascore', 'Star_1', 'Star_2', 'Star_3', 'Star_4',
           'Votes', 'Genre1', 'Genre2', 'Genre3', 'Gross collection',
           'Popularity', 'Certification']

# Iterate over each column
for column in columns:
    # Check if the column contains string values
    if isinstance(merge_df[column][0], str):
        # Skip columns with string values
        continue
    else:
        # Calculate quartiles
        quartiles = percentile(merge_df[column], [25, 75])
        # Calculate lower and upper fences
        lower_fence = quartiles[0] - (1.5 * (quartiles[1] - quartiles[0]))
        upper_fence = quartiles[1] + (1.5 * (quartiles[1] - quartiles[0]))
        # Apply outlier treatment using lambda function
        merge_df[column] = merge_df[column].apply(lambda x: upper_fence if x > upper_fence else (lower_fence if x < lower_fence else x))
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Create the models
models = [
    LinearRegression(),
    GradientBoostingRegressor(),
    RandomForestRegressor(),
    DecisionTreeRegressor(),
    KNeighborsRegressor()
]

# Initialize variables to track the best model and its difference
best_model = None
best_difference = np.inf

# Train and evaluate each model
for model in models:
    # Fit the model
    model.fit(x_train, y_train)
    
    # Predict on test data
    y_pred = model.predict(x_test)
    
    # Compute the evaluation metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    
    # Print the evaluation metrics
    print(f"Model: {type(model).__name__}")
    print(f"R2 Score: {r2:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print("-------------------------------------------")
    
    # Compute the difference between the R2 score and cross-validation R2 score
    cv_scores = cross_val_score(model, x, y, cv=5, scoring='r2')
    cv_score = np.mean(cv_scores)
    difference = abs(r2 - cv_score)
    
    # Update the best model if the difference is lower
    if difference < best_difference:
        best_model = model
        best_difference = difference

# Print the best model and its difference
print("Best Model:")
print(type(best_model).__name__)
print(f"Difference: {best_difference:.4f}")
```
